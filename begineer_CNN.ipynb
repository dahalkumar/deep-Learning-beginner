```python
import cv2
import numpy as np
```

```python
def convolution_filter_demo(image_source,kernel):
    kernel_length = len(kernel)
    #output image size
    row = image_source.shape[0] - kernel_length+1
    col = image_source.shape[1] - kernel_length+1
    #create all pixel zero except row and col
    op_image = np.zeros(shape = (row,col))
    print(row,col)
    #now for every row and col
    for i in range(row):
        for j in range(col):
            current_matrix_image = image_source[i:i+kernel_length, j:j+kernel_length]
            #multiply kernel with current_matrix_image
            multiply = np.abs(np.sum(np.sum(current_matrix_image*kernel)))
            op_image[i,j] = multiply
    return op_image        
```

```python
img = cv2.imread("cat.jpg",cv2.IMREAD_GRAYSCALE)
channel = None
if len(img.shape) == 2:
    height,width = img.shape
elif len(img.shape) == 3:
    height,width,channel = img.shape
print(height,width,channel)    
```

```python
for n in range(5):
      k = [[-1, -1, 1], [-1, n, -1], [0, -1, 0]] 
      filtered_image = convolution_filter_demo(img, k)
      plt.imshow(filtered_image)
      plt.show()
      print("------------------------------------------")
      cv2.waitKey(0)
```
